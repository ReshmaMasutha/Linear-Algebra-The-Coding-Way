**2.Adjacency matrix:**
Create a 4x4 adjacency matrix A, where each entry Aij is 1 if there is directed edge from pagei to page j, and otherwise.



**3.Transition probability matrix:**


* Calculate the number of outgoing links from each page.
* Divide each row of the adjacency matrix by the corresponding number of outgoing links to obtain the transition probability matrix P.
* Each element in the matrix represents the probability of transitioning from the page represented by the jth column to the page represented by the ith row.
