**Solution for the problem:**

The first step to determine whether a given vector is an eigenvector of a matrix A involves performing a matrix-vector multiplication between the matrix 𝐴 and the given vector. Let's denote the given vector as v.

So, if v is a potential eigenvector of A corresponding to an eigenvalue λ, then:

   Av=λv

If after performing the matrix-vector multiplication you obtain a scalar multiple of v, then v is an eigenvector of A with eigenvalue λ. Otherwise,v is not an eigenvector of A.


**To Prove The Eigen Vectors Visually We Need To Follow This Steps:**

**1.Plot the original vector:** Start by plotting the given vector in a coordinate system. This gives you a visual representation of the direction and magnitude of the vector.

**2.Apply the transformation by the matrix:** Multiply the matrix A by the vector. This will give you a new vector, which represents the result of transforming the original vector by the matrx A.

**3.Compare the transformed vector to the original vector:**  If the transformed vector is a scalar multiple of the original vector, meaning it lies on the same line but possibly stretched or shrunk, then the original vector is an eigenvector of the matrix. The scalar by which it's stretched or shrunk corresponds to the eigenvalue associated with that eigenvector.

**4.Repeat for different eigenvalues:** If you have multiple eigenvalues and corresponding eigenvectors, you can repeat this process for each eigenvector to visually verify them.


```python
Program :
# Import the necessay libraries
import numpy as np
import matplotlib.pyplot as plt
```

Note:'numpy'  is a library for numerical operations in Python, particularly useful for handling arrays and matrices.

'matplotlib.pyplot'  is a plotting library used for visualizing data.

```python
# Define Matrix A
A = np.array([[1, 6], [5, 2]])
```

Note:This line defines the matrix A using NumPy's array function. It's a 2x2 matrix as provided.

```python
# Define vector u and v
u = np.array([6, -5])
v = np.array([3, -2])
```

Note:These lines define the vectors u and v as NumPy arrays.


```python
# Calculate A*u and A*v
Au = np.dot(A, u)
Av = np.dot(A, v)

print(Au)
print(Av)
```

Note:These lines calculate the matrix-vector multiplication Au and Av using NumPy's 'dot' function.

```python
# Setting up the figure and specifying the figure size
plt.figure(figsize=(8, 6))
# Plotting vector u from origin to its coordinates,labeled 'u' in red
plt.plot([0, u[0]], [0, u[1]], 'r', label='u')
# Plotting vector v from origin to its coordinates,labeled 'v' in capri blue
plt.plot([0, v[0]], [0, v[1]], 'c', label='v')
# Plotting  the transformed vector Au from origin to its coordinates,labeled 'Au' in green
plt.plot([0, Au[0]], [0, Au[1]], 'g', label='A * u')
# Plotting  the transformed vector Av from origin to its coordinates,labeled 'Av' in majorelle blue
plt.plot([0, Av[0]], [0, Av[1]], 'm', label='A * v')
# Adding a grid to the plot for better visibility of vector positions
plt.grid(True)
# Setting labels for the X and Y axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# Setting a title for the plot
plt.title('Visualization of Eigenvectors under A matrix')
# Adding a legend to the plot to identify each vector
plt.legend()
plt.axis('equal')
# Displaying the plot
plt.show()
```


Note:
* These lines create a plot using Matplotlib.

* Each' plt.plot' function call plots a vector. For example, plt.plot([0, u[0]], [0, u[1]], 'b', label='u') plots vector u in blue color ('b').

* The other vectors Au and Av are also plotted similarly.

* 'plt.xlabel', 'plt.ylabel', and 'plt.title' set the labels and title for the plot.


* 'plt.legend()' adds a legend to the plot.   


* 'plt.grid(True)' adds grid lines to the plot.   




* 'plt.axis('equal')' ensures that the aspect ratio of the plot is equal, so the vectors appear in their correct proportions.
* 'plt.show()' displays the plot.
